```python
values = []
cumulative_values = []
x_axis = []
target = data[data.columns[-1]].sum()

for i in range(-8,-1):
    values.append(data[data.columns[i]].sum())
    x_axis.append(' '.join(data.columns[i].split()[-2:-1]))
    
cumulative_values.append(values[0])
for i in range(1,len(values)):
    cumulative_values.append(values[i] + cumulative_values[i-1])
    
print(values)
print(cumulative_values, target)
print(x_axis)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel('United-Way-Code-for-Good-combined-performance-data-v2.xlsx')
for i in range(-8,0):
    data[data.columns[i]] = data[data.columns[i]].fillna(0)

def generate_plots(STRATEGY_NAME, METRIC_NAME):
    frame = data[(data['Strategy_Name']==STRATEGY_NAME)]

    metric_names = list(set(frame['Outcome_Indicator_Name']))
    metric_sums = []
    metric_targets =[]
    for metric_name in metric_names:
        metric_sum = 0
        metric_frame = frame[frame['Outcome_Indicator_Name']==metric_name]
        for i in range(-8,-1):
            metric_sum += metric_frame[metric_frame.columns[i]].sum()
        metric_sums.append(metric_sum)
        metric_targets.append(metric_frame[metric_frame.columns[-1]].sum())
    
    N = len(metric_names)
    ind = np.arange(N) 
    width = 0.35       
    plt.bar(ind, metric_sums, width, label='Actual to Date')
    plt.bar(ind + width, metric_targets, width,
        label='Annual Target')
    
    plt.ylabel('Value')
    
    #plt.xticks(ind + width / 2, metric_names)
    plt.legend(loc='best')
    plt.show()
    
    frame = data[(data['Strategy_Name']==STRATEGY_NAME)&(data['Outcome_Indicator_Name']==METRIC_NAME)]
    
    values = []
    cumulative_values = []
    x_axis = []
    target = frame[frame.columns[-1]].sum()
    
    for i in range(-8,-1):
        values.append(frame[frame.columns[i]].sum())
        x_axis.append(' '.join(frame.columns[i].split()[-2:-1]))
        
    cumulative_values.append(values[0])
    for i in range(1,len(values)):
        cumulative_values.append(values[i] + cumulative_values[i-1])
    
    plt.hlines(target,0,len(cumulative_values)-1, 'r', 'dotted')
    plt.plot(cumulative_values, '-bo')
    plt.xticks(np.arange(len(x_axis)), x_axis)
    plt.show()
    
```
