```python
values = []
cumulative_values = []
x_axis = []
target = data[data.columns[-1]].sum()

for i in range(-8,-1):
    values.append(data[data.columns[i]].sum())
    x_axis.append(' '.join(data.columns[i].split()[-2:-1]))
    
cumulative_values.append(values[0])
for i in range(1,len(values)):
    cumulative_values.append(values[i] + cumulative_values[i-1])
    
print(values)
print(cumulative_values, target)
print(x_axis)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.hlines(target,0,len(cumulative_values)-1, 'r', 'dotted')
plt.plot(cumulative_values, '-bo')
locs, labels = plt.xticks()            # Get locations and labels
plt.xticks(np.arange(len(x_axis)), x_axis)
plt.show()
```
